# Type 1 Analysis

## Jumlah data tiap tipe 1

```python
# Data dan plot diagram batang
Type1_counts = df['Type 1'].value_counts()
Type1_counts.plot(kind='bar', xlabel='Type 1', ylabel='Count', rot=0, figsize=(15, 6))

# Menambahkan label pada setiap bar
for i, count in enumerate(Type1_counts):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.show()
```

## Ada berapa legendary tiap tipe

```python
tipe_pokemon = df['Type 1'].unique()
tipe_pokemon
```

```python
import matplotlib.pyplot as plt


hasil_count_T1 = {}

for tipe in tipe_pokemon:
    type1_legendary_true = df.loc[(df['Legendary'] == True) & (df['Type 1'] == tipe)].shape[0]
    type1_legendary_false = df.loc[(df['Legendary'] == False) & (df['Type 1'] == tipe)].shape[0]
    hasil_count_T1[tipe] = [type1_legendary_true, type1_legendary_false]

y1 = []
y2 = []

for tipe in tipe_pokemon:
    y1.append(hasil_count_T1[tipe][0])
    y2.append(hasil_count_T1[tipe][1])
```

```python
print(y1)
print(y2)
```

```python
hasil_count_T1
```

```python
tipe_pokemon[0]
```

```python
for i in range(18):
  print(i)
  print(f'Nama Pokemon : {tipe_pokemon[i]}')
  print(f'Jumlah Legendary : {y1[i]}')
  print(f'Jumlah bukan Legendary : {y2[i]}')
```

```python
fig, ax = plt.subplots(figsize=(12, 8))


# y1.sort()
# y2.sort()

y1 = np.array(y1)
y2 = np.array(y2)

# Menggunakan argsort untuk mendapatkan indeks yang mengurutkan y1
sorted_indices_T1 = np.argsort(y1)

# Menggunakan indeks untuk mengurutkan array tipe_pokemon, y1, dan y2
tipe_pokemon_sorted = tipe_pokemon[sorted_indices_T1]
y1_sorted = y1[sorted_indices_T1]
y2_sorted = y2[sorted_indices_T1]

# plot bars in stack manner with horizontal bars
ax.barh(tipe_pokemon_sorted, y1_sorted, color='#f84b4b', label='Legendary True')
ax.barh(tipe_pokemon_sorted, y2_sorted, left=y1_sorted, color='#1D76DB', label='Legendary False')

# Menambahkan legenda
ax.legend()

plt.ylabel("Type 1",  fontsize=14)
plt.xlabel("Count", fontsize=14)

plt.show()
```

## Ada berapa legendary tiap generation

```python
tipe_generation = df["Generation"].unique()
tipe_generation
```

```python


hasil_count_gen = {}

for tipe in tipe_generation:
    generation_legendary_true = df.loc[(df['Legendary'] == True) & (df['Generation'] == tipe)].shape[0]
    generation_legendary_false = df.loc[(df['Legendary'] == False) & (df['Generation'] == tipe)].shape[0]
    hasil_count_gen[tipe] = [generation_legendary_true, generation_legendary_false]

gen_y1 = []
gen_y2 = []

for tipe in tipe_generation:
    gen_y1.append(hasil_count_gen[tipe][0])
    gen_y2.append(hasil_count_gen[tipe][1])
```

```python
for i in range(len(tipe_generation)):
  print(f'Generasi : {tipe_generation[i]}')
  print(f'Jumlah Legendary : {gen_y1[i]}')
  print(f'Jumlah bukan Legendary : {gen_y2[i]}')
```

```python
gen_y1
```

```python
tipe_generation = np.array(['1=A','2=B','3=C','4=D','5=E','6=F'])
tipe_generation
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
# tipe_generation = str(tipe_generation)
# Menggunakan argsort untuk mendapatkan indeks yang mengurutkan gen_y1

gen_y1 = np.array(gen_y1)
gen_y2 = np.array(gen_y2)
sorted_indices = np.argsort(gen_y1)

# Menggunakan indeks untuk mengurutkan array tipe_generation, gen_y1, dan gen_y2
tipe_generation_sorted = tipe_generation[sorted_indices]
gen_y1_sorted = gen_y1[sorted_indices]
gen_y2_sorted = gen_y2[sorted_indices]

# Plot bars in stack manner with horizontal bars
ax.barh(tipe_generation_sorted, gen_y1_sorted, color='#f84b4b', label='Legendary True')
ax.barh(tipe_generation_sorted, gen_y2_sorted, left=gen_y1_sorted, color='#1D76DB', label='Legendary False')

# Menambahkan legenda
ax.legend()
plt.xlabel("Count", fontsize=14)
plt.ylabel("Generation", fontsize=14)

plt.show()
```
